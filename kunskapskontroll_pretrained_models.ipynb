```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
# img_path = 'zebra.jpg'
# img_path = 'linjal.jpg'
img_path = 'tennisboll.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?

Using pretrained AI models saves time and computational resources compared to training from scratch, which demands large amounts of data and processing power. Pretrained models, trained on big and varied datasets, cut out the long initial training phase, letting developers and researchers launch AI solutions quicker. They typically achieve higher accuracy, particularly in tasks with limited data availability. By leveraging their existing understanding of complex patterns and features from large datasets, these models can effectively apply this knowledge to new, similar tasks. This prior knowledge contributes to improved performance compared to models trained from scratch on smaller datasets.
